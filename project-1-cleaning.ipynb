Now let's try to clean up our data by looking at the goal of our questions. We want to know if students attented any CCM events so we have to create a new column that will look at all columns related to events and fill in the new column accordingly.

```python
#Let's create a new column that will tell us if a student attended ANY of the CCM events prior to enrolling at CCM.
event = ['open_house', 'instant_decision', 'campus_info', 'virtual_info',
       'women_whodare', 'college_fair', 'sharetime_info', 'challenger',
       'cybersc_prog', 'hs_info', 'hscampus_visit', 'ind_campus_visit',
       'worforce_class', 'corporate_training', 'teen_arts', 'summer_camp',
       'hour_code'] 
computing_selected['attended_event'] = 'No'

#I used stackoverflow to learn about iterrows: https://stackoverflow.com/questions/16476924/how-to-iterate-over-rows-in-a-dataframe-in-pandas
#I didn't know about .at[] either, I kind of figured it out while reading different websites, here's one
# of them: https://www.geeksforgeeks.org/python-pandas-dataframe-at/
#If in any of the columns in event is Yes, then change the new column to say Yes to indicate the student 
#attented an event at CCM prior to enrollment.
for index, row in computing_selected.iterrows():
    for column in event:
        if row[column] == 'Yes':
            computing_selected.at[index, 'attended_event'] = 'Yes'
            break
```

Now let's do the same thing to see if students received information about the Technology department at CCM.

```python
#Let's create a new column that will tell us if a student received ANY information about the technology 
#department at CCM prior to enrolling at CCM.
info = ['hs_councelor', 'teacher', 'IT_website',
       'admissions', 'CCM_advisor', 'employer', 'ccmworforce', 'njworkforce',
       'other'] 
computing_selected['received_info'] = 'No'

for index, row in computing_selected.iterrows():
    for column in info:
        if row[column] == 'Yes':
            computing_selected.at[index, 'received_info'] = 'Yes'
            break
```

```python
#Now that we have the column that says if a student attented an event, we don't really need each individual 
#event(our questions do not ask the distribution) so lets subset again to remove exclude them from our data.
columns_staying = ['Major', 'hs_councelor', 'teacher', 'IT_website',
       'admissions', 'CCM_advisor', 'employer', 'ccmworforce', 'njworkforce',
       'other', 'gender', 'race/ethnicity', 'age', 'intereset_rating',
       'first_major', 'attended_event', 'received_info']

computing_selected = computing_selected[columns_staying]
computing_selected.columns #Looks like we're left with the columns we need to answer our questions.
```
